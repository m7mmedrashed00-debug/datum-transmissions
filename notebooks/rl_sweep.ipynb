```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent / 'src'))
import numpy as np, matplotlib.pyplot as plt
from datum import rl_coil as m
```

```python
# Temperature sweep: resistance and steady current at fixed 12 V
T = np.linspace(20, 140, 61)
R = np.array([m.resistance(5.0, t) for t in T])
fig, ax = plt.subplots(1, 2, figsize=(10, 3.2))
ax[0].plot(T, R); ax[0].set_xlabel('T (degC)'); ax[0].set_ylabel('R (ohm)')
ax[1].plot(T, 12.0 / R); ax[1].set_xlabel('T (degC)'); ax[1].set_ylabel('steady I at 12 V (A)')
fig.suptitle('Hot coil: resistance rises, constant-voltage current droops'); plt.tight_layout()
```

```python
# Step response at cold vs hot resistance — tau = L/R changes with temperature
t = np.linspace(0, 0.02, 400)
for temp in (20, 120):
    r = m.resistance(5.0, temp)
    plt.plot(t * 1e3, [m.rl_step_current(x, 12.0, r, 0.012) for x in t], label=f'{temp} degC')
plt.xlabel('t (ms)'); plt.ylabel('i (A)'); plt.legend(); plt.title('RL step response')
```

```python
# Voltage compliance check (Ch 8 worked example)
r120 = m.resistance(5.0, 120.0)
print(f'R(120C) = {r120:.3f} ohm')
print(f'V_required = {m.v_required(1.40, 700.0, r120, 0.012, 0.8):.2f} V  (supply must exceed this)')
```
